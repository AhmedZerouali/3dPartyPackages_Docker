# Let's get the latest available versions of installed packages

```python
packages = pd.read_csv('../data/cleaned_installed_packages.csv', dtype=str)
packages.package.fillna('nan', inplace=True)
packages.shape
```

```python
packages[['base','package']].drop_duplicates().groupby('base').count()
```

```python
def download_url(url):
    ''' Download an url and returns the content'''
    
    response = requests.get(url)
    webContent  = response.json()
    return webContent
```

```python
# Let's start with npm (I hope it is easy for the rest :) )
names = []
versions = []
dates = []

mask = list(packages.query('base == "node"').package)
mask = set(mask)

for name in tqdm(mask, total=len(mask)):
    url = 'https://registry.npmjs.org/'+name
    try:
        data  = download_url(url)
        for version in data['time']:
            try:
                if version== 'modified' or version=='created':
                    continue
                names.append(name)
                versions.append(version)
                dates.append(data['time'][version])
            except: 
                continue
    except:
        print(name,2)
        continue
npm = pd.DataFrame({'package': names, 'version': versions, 'date': dates, 'base':'node'})
```

```python
npm.to_csv('../data/npm_versions.csv', index=False)
```

```python
# Now let's try with python

names = []
versions = []
dates = []

mask = (list(packages.query('base == "python2"').package) +
        list(packages.query('base == "python3"').package))
mask = set(mask)

for name in tqdm(mask, total=len(mask)):
    url = 'https://pypi.org/pypi/'+name+'/json'
    try:
        data  = download_url(url)
        for version in data['releases']:  
            try: 
                date = data['releases'][version][0]['upload_time']
                names.append(name)
                versions.append(version)
                dates.append(date)
            except:
                continue
    except:
        print(name,2)
        continue

    
python = pd.DataFrame({'package': names, 'version': versions, 'date': dates, 'base':'python'})
```

```python
python.to_csv('../data/pypi_versions.csv', index=False)
```

```python
# Now let's try with ruby

names = []
versions = []
dates = []

mask = list(packages.query('base == "ruby"').package)
mask = set(mask)

for name in tqdm(mask, total=len(mask)):
    url = 'https://rubygems.org/api/v1/versions/'+name+'.json'
    try:
        data  = download_url(url)
        for element in data:
            try:
                names.append(name)
                versions.append(element['number'])
                dates.append(element['created_at'])
            except:
                continue
    except:
        print(name)
        continue
    
ruby = pd.DataFrame({'package': names, 'version': versions, 'date': dates, 'base':'ruby'})
```

```python
ruby.to_csv('../data/ruby_versions.csv', index=False)
```

```python
packages = pd.read_csv('../data/lag_release.csv', 
                       usecols = ['package','base','core'], dtype=str)
packages.package.fillna('nan', inplace=True)
packages.drop_duplicates(inplace=True)
packages.shape
```

```python
paqs = []
paqs2 = []
packs = packages[['base','package']].drop_duplicates()
for i, row in tqdm(enumerate(packs.iterrows()), total = len(packs)):
    package = row[1]['package']
    if row[1]['base'] == "node":
        url = 'https://registry.npmjs.org/'+package
        try:
            data  = download_url(url)
            description = data['description']
        except:
            description= ""
            
        try:
            summary = data['keywords']
        except:
            summary = ""  
            
    elif row[1]['base'] == "ruby":
        url = 'https://rubygems.org/api/v1/gems/'+package+'.json'
        try:
            data  = download_url(url)
            description = data['info']
        except:
            description= ""
            
        summary = ""  
    
    else:
        url = 'https://pypi.org/pypi/'+package+'/json'
        try:
            data  = download_url(url)
            description = data['info']['description']
        except:
            description= ""
            
        try:
            summary = data['info']['summary']
        except:
            summary = ""  
            
    #description = ' '.join(set(description.split()))
    paqs.append(description)
    paqs2.append(summary)
        
packs['info'] = paqs
packs['summary'] = paqs2
```

```python
# Keyword for only node
all_keys= []
node_packages = packages.query('base == "node" and core=="False"')
for i, row in tqdm(enumerate(node_packages.iterrows()), total = len(node_packages)):
    package = row[1]['package']
    if row[1]['base'] == "node":
        url = 'https://registry.npmjs.org/'+package
        try:
            data  = download_url(url)
            for x in data['versions']:
                keywords = data['versions'][x]['keywords']
                all_keys = all_keys + keywords
        except:
            pass
        
```

```python
packages['info2'] = packages['info'].apply(lambda x: ' '.join(set(x.split())).lower())
```

```python
packages['info2'] = packages['info2'].apply(lambda x: x.replace('the',''))
packages['info2'] = packages.apply(lambda d: d['info2'].replace(d['package'].lower(),''), axis=1)
packages['info2'] = packages['info2'].apply(lambda x: x.replace('javascript',''))
packages['info2'] = packages['info2'].apply(lambda x: x.replace('node',''))
packages['info2'] = packages['info2'].apply(lambda x: x.replace('library',''))
packages['info2'] = packages['info2'].apply(lambda x: x.replace('module',''))
packages['info2'] = packages['info2'].apply(lambda x: x.replace('package',''))
packages['info2'] = packages['info2'].apply(lambda x: x.replace('ruby',''))
packages['info2'] = packages['info2'].apply(lambda x: x.replace('rubygems',''))
packages['info2'] = packages['info2'].apply(lambda x: x.replace('python',''))
packages['info2'] = packages['info2'].apply(lambda x: x.replace('that',''))
packages['info2'] = packages['info2'].apply(lambda x: x.replace('with',''))
packages['info2'] = packages['info2'].apply(lambda x: x.replace('from',''))
```

```python
from os import path
from wordcloud import WordCloud
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

words = {}
for word in all_keys:
    try:
        words[word] = words[word]+1
    except:
        words[word] = 1

wordcloud = WordCloud( width = 1800, height = 900, background_color='white').generate_from_frequencies(words)
plt.figure(figsize=(21,12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
